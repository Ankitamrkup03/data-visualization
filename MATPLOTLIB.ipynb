```python
import pandas as pb
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
```

```python
df=pb.read_csv("gas_prices.csv")
df
```

LINE GRAPH

```python
#plot the data in the graph
plt.plot(df.Year,df.Australia)
plt.plot(df.Year,df.Canada)
plt.plot(df.Year,df['South Korea'])
plt.show()
```

```python
#plot the graph with dots and change the color
plt.plot(df.Year,df.Australia,'g.-')
plt.plot(df.Year,df.Canada,'b*-')
plt.plot(df.Year,df['South Korea'],'r+-')
```

```python
plt.figure(figsize=(8,5)) #increase the size of the graph

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})  #give the title of the graph and 
                                                                                            #font size

plt.plot(df.Year,df.Australia,'g.-',label='Australia')
plt.plot(df.Year,df.Canada,'b*-',label='canada')
plt.plot(df.Year,df['South Korea'],'r+-',label='South Korea')

plt.xlabel('year',fontdict={'fontweight':'bold', 'fontsize': 18})
plt.ylabel('USA dollar',fontdict={'fontweight':'bold', 'fontsize': 18})

plt.legend()    #show the legend in the graph

plt.show()
```

```python
bins=[1,2,3]
plt.hist(df.Australia, bins=bins, color='#abcdef')
plt.show()
```
